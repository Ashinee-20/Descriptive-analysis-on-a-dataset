```python
import math
import numpy as np
import pandas as pd
import statistics
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('bodyPerformance.csv')
df.head()
```

```python
df.dtypes
```

```python
num_data = df.select_dtypes(exclude = 'object')
cat_data = df.select_dtypes(include = 'object')
```

```python
#mean
print(df.mean())
print("The mean of age column is: ", df['age'].mean())
```

```python
#mean median mode
print("Mean: ")
print(df.mean())
print("\n")
print("Median: ")
print(df.median())
print("\n")
print("Mode: ")
print(df.mode())
print("\n")
```

```python
statistics.variance(df['age'])
```

```python
statistics.stdev(df['weight_kg'])
```

```python
range = df.age.max() - df.age.min()
print("Range of the age column is: ", range)
```

```python
df.skew()
```

```python
print(df.age.skew())
print("since skewness is above 0, so it is positive skewness")
df.age.plot(kind = "density")
```

```python
print(df['sit-ups counts'].skew())
print("since skewness is less than 0, so it is negative skewness")
df['sit-ups counts'].plot(kind = "density")
```

```python
print("Density of all features: ")
df.plot.density(figsize = (15,8))
```

```python
df.kurt()
print(df.columns)
```

```python
for col in num_data:
  if(df[col].kurt()>3):
    print(col + " is leptokrutic")
  elif(df[col].kurt()<3):
    print(col + " is platykrutic")
  else:
    print(col + " is mesokrutic")
```
